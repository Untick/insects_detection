```python
cap = cv2.VideoCapture("/content/drive/MyDrive/Cabbage/origcam2.CUT.01'23.000-01'33.000.mp4")
fps = cap.get(cv2.CAP_PROP_FPS)
frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
print(f'Файл {cap} содержит всего {frames} кадров ({fps} кадров/с).')
file_count = 0


while(cap.isOpened()):
    ret, frame = cap.read()
    if ret:
        # Блюр
        img = cv2.medianBlur(frame, 33)
        # Перевод в градации серого
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        # Отрисовка граней
        canny_image = cv2.Canny(gray, 10, 50)
        # Возвтращаем размерность
        canny_3 = cv2.cvtColor(canny_image, cv2.COLOR_GRAY2BGR)
        # Счетчик обработанных кадров
        file_count += 1
        print(f'Кадр {file_count}')
        # Наложение изображений с гранями на оригинальные цветные кадры
        img = cv2.addWeighted(canny_3, 1, frame, 0.7, 0.0)
        # Сохранение кадров на диск
        writefile = f'/content/drive/MyDrive/Cabbage/Images/{file_count}.jpg'
        cv2.imwrite(writefile, img)
    else:
      break
cv2.destroyAllWindows()
cap.release()
```
